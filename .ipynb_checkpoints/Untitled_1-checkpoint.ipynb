```python
import os
import random
import glob
import rasterio
from rasterio.transform import from_origin
from rasterio.features import geometry_mask
from shapely.geometry import Point
import geopandas as gpd
import numpy as np
import fiona

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 랜덤포인트 개수
num_points = 500
```

```python
# 환경변수 폴더 경로
env_data_path = "env_data/test_raster"
# 기준 래스터 파일 경로
```

```python
# 기준 래스터 파일 경로
ref_file_path = 'env_data/test_raster/aspect.tif'

# 랜덤포인트 shapefile 경로
sinkhole_points_path = r'env_data/sinkhole_loc/sinkhole_loc_219.shp'
random_points_path = f'env_data/sinkhole_loc/random_points_{num_points}.shp'

# 환경변수 래스터 파일 경로 목록
env_file_paths = glob.glob(f"env_data/test_raster/*.tif")  # 필요에 따라 추가 경로 포함
```

# 1. 랜덤포인트 shp파일 생성

```python
# Open the raster file using rasterio
with rasterio.open(ref_file_path) as src:
    # Read the raster data
    data = src.read(1)
    nodata_value = src.nodata
    # print(nodata_value)
    
    # Get valid bounds based on non-NaN values
    valid_bounds = src.bounds
    
    # Generate random points within the valid bounds
    random_points = []
    for _ in range(num_points):
        while True:
            x = random.uniform(valid_bounds.left, valid_bounds.right)
            y = random.uniform(valid_bounds.bottom, valid_bounds.top)
            row, col = src.index(x, y)
            if not (row < 0 or col < 0 or row >= data.shape[0] or col >= data.shape[1] or data[row, col] == nodata_value):
                random_points.append(Point(x, y))
                break

# Create a GeoDataFrame from the random points
gdf = gpd.GeoDataFrame(geometry=random_points, crs=src.crs)

# Save the GeoDataFrame to a shapefile
gdf.to_file(random_points_path)
print(f"{random_points_path} 파일에 랜덤 포인트 {num_points}개가 생성되었습니다.")
```

# 2. shp파일의 위치데이터의 환경변수 값 추출 함수

```python
def  extract_value_by_point(shapefile_path):
    # shapefile을 GeoDataFrame으로 읽기
    gdf = gpd.read_file(shapefile_path, encoding='euc-kr')
    
    # 데이터를 저장할 데이터프레임 생성
    data_df = pd.DataFrame()
    
    # 모든 래스터 파일에 대해 루프
    for raster_path in raster_file_paths:
        with rasterio.open(raster_path) as src:
    
            col_name = os.path.basename(raster_path).split('.')[0]
            # 래스터 데이터를 읽어옴
            data = src.read(1)
            
            # 각 포인트 위치에서 값을 추출하고 데이터프레임에 추가
            for index, row in gdf.iterrows():
                point = row['geometry']
                x, y = point.x, point.y
                row_value = data[src.index(x, y)]
                data_df.at[index, col_name] = row_value
    return data_df
    
```

```python
sinkhole_data = extract_value_by_point(sinkhole_points_path)
sinkhole_data['label'] = 1
sinkhole_data.head()
```

```python
# 각 열의 최소값 찾기
min_values = sinkhole_data.min()
min_values
```

```python
# 최소값과 동일한 값을 갖는 항목에 np.nan 할당
for column in sinkhole_data.columns:
    if column != "label":
        sinkhole_data.loc[sinkhole_data[column] == min_values[column], column] = np.nan
```

```python
# 이상치 처리된 데이터 : sinkhole-preprocessing.csv
# Vif 처리된 데이된 : sinkhole-preprocessing_VIF.csv
model_df = pd.read_csv(f'results/model_test_data.csv')
```

```python
TARGET = "label"
RANDOM_STATE = 2023
```

```python
X_df = model_df.drop(TARGET , axis =1 )
y_df = model_df[TARGET]
X_train , X_test , y_train , y_test = train_test_split(X_df , 
                                                       y_df, 
                                                       random_state = RANDOM_STATE,
                                                       train_size =0.8)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay

def plot_confusion_matrix(model, X_test, y_test):
    disp = ConfusionMatrixDisplay.from_estimator(model,
                                                 X_test, y_test,
                                                 cmap=plt.cm.Blues)
def plot_roc_curve(model, X_test, y_test):
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.show()
    
def evaluate(model, y_pred):
    plot_confusion_matrix(model, X_test, y_test)
    print(classification_report(y_test, y_pred))
    plot_roc_curve(model, X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
lgr.fit(X_train, y_train)

lgr_preds = lgr.predict(X_test)
```

```python
evaluate(lgr, lgr_preds)
```

```python
def save_predict_map(output_file, map_data):
    with rasterio.open(raster_paths[0]) as src:
        
        # data = src.read()
        data = map_data
    
         # 출럭 파일 생성
        with rasterio.open(output_file, 'w', driver='GTiff',
                           height=src.height, width=src.width,
                           count=src.count, dtype=data.dtype,
                           crs=src.crs, transform=src.transform) as dst:
            # 데이터 쓰기
            dst.write(data, 1)
    
    print(f"{output_file}에 데이터가 저장되었습니다.")        
```

```python
def make_predict_map(model):
    map_class = model.predict_proba(map_data_scaled)
    map_data = map_class[:, -1].reshape(raster_size[0], raster_size[1])
    print(map_data.shape)

    model_name = model.__class__.__name__

    output_file = f"results/{model_name}_predict_map.tif"
    save_predict_map(output_file, map_data)    
    plt.imshow(map_data, cmap='plasma')
    plt.colorbar()
```
